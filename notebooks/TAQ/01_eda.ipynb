## Observations of WRDS TAQ derived data

The data sample runs from June, 1st 2018 to August, 13 2020

---

```python
# load libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
df_main = pd.read_csv("../../data/taq/taq.csv")
```

```python
df_main.head()
```

```python
# number of unique symbols
print(df_main["symbol"].nunique())
```

```python
# intersection of unique symbols in TAQ data and unique tickers in Barber et al. (2022)
with open("../../data/replication/unique_tickers_main_pseudo.txt", "r") as f:
    unique_tickers_main = set(line.strip() for line in f)

unique_symbols_taq = set(df_main["symbol"].unique())
overlap_symbols = unique_tickers_main.intersection(unique_symbols_taq)
print(f"Number of unique tickers in Barber et al. (2022) main data: {len(unique_tickers_main)}")
print(f"Number of unique symbols in TAQ data: {len(unique_symbols_taq)}")
print(f"Number of overlapping unique symbols: {len(overlap_symbols)}")
```

```python
# missing values in TAQ data
reports = df_main.isna().sum()
for column, count in reports.items():
    print(f"Column '{column}' has {count} missing values.")
```

```python
# length of data set
print(f"Number of rows in TAQ data: {len(df_main)}")
```
