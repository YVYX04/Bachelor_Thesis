```python
# libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import seaborn as sns
from datetime import datetime
```

I then load the data.

```python
attention = pd.read_csv("../../../data/processed/herding_events_full_attention.csv")
base = pd.read_csv("../../../data/processed/herding_events_full.csv")

# parses dates
attention["date"] = pd.to_datetime(attention["date"], format="%Y-%m-%d")
base["date"] = pd.to_datetime(base["date"], format="%Y-%m-%d")

# size
attention["size"] = attention["prc"].abs() * attention["shrout"] / 1000  # in millions
base["size"] = base["prc"].abs() * base["shrout"] / 1000  # in millions
```

```python
# percentage of nan values for asvi
asvi_nan_attention = attention["asvi"].isna().mean() * 100
asvi_nan_base = base["asvi"].isna().mean() * 100
print(f"Percentage of NaN values for ASVI in attention dataset: {asvi_nan_attention:.2f}%")
print(f"Percentage of NaN values for ASVI in base dataset: {asvi_nan_base:.2f}%")
```

```python
# percentage of rh_herd 
p_herd = np.sum(attention['rh_herd'] == 1) / len(attention) * 100
print(f"Percentage of rh_herd events in attention dataset: {p_herd:.2f}%")
```
