## Google Trends Data

---

In this notebook, I illustrate what Google Trends raw data can look like.

```python
# libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

# data
df = pd.read_csv(
    "../../../data/raw/iran_oil.csv"
)

# head
df.head()
```

```python
# rename 'Time' to 'date' and convert to datetime
df.rename(columns={"Time": "date"}, inplace=True)
df["date"] = pd.to_datetime(df["date"])

# sort by date
df.sort_values("date", inplace=True)
```

Once the data set is ready, I realize the plot.

```python
# custom HSG colors
green = (24/255, 127/255, 56/255)
beige = (188/255, 169/255, 128/255)
red = (229/255, 110/255, 108/255)
```

```python
# plot
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(
        df["date"],
        df["Iran war"],
        color="blue",
        linewidth=1.5,
        label="Iran war"
    )

ax.plot(
        df["date"],
        df["oil price"],
        color=green,
        linewidth=1.5,
        label="Oil price"
    )

# vertical line for Feb 28, 2026
ax.axvline(
    pd.Timestamp("2026-02-28"),
    color=red,
    linewidth=1.5,
    alpha=0.9
)

ax.set_title("Google Trends Search Volume Index (SVI) for 'Iran war' and 'oil price'", fontsize=16, loc = "left",
             family="Times New Roman", pad=20)
ax.set_ylabel("Search Volume Index (SVI)", fontsize=12, family="Times New Roman")
ax.legend()
#ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.setp(ax.get_xticklabels(), rotation=45, ha="right") 
ax.grid(axis="both", alpha=0.25, linewidth=0.8)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
fig.tight_layout()
plt.savefig(
    "../../../figures/google_trends_iran_war_oil_price.pdf",
    bbox_inches="tight"
)
plt.show()
```
