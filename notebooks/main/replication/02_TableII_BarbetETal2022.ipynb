## 1. Libraries & Data

I first load the relevant libraries and data.

```python
# libs 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import seaborn as sns
from datetime import datetime

# data
rh_herd = pd.read_csv("../../../data/processed/herding_events_only.csv")

# parses dates
rh_herd["date"] = pd.to_datetime(rh_herd["date"], format="%Y-%m-%d")

# size
rh_herd["size"] = rh_herd["prc"].abs() * rh_herd["shrout"] / 1000  # in millions

# mean size
mean_size = rh_herd["size"].mean()
print(f"Mean size: {mean_size:.2f}")
```

I print the summary statistics.

```python
# summary stats
summary_stats = rh_herd[["users_close", "users_last", "userchg", "userratio", "prc", "ret", "size", "daily_buys", "daily_sells", "net_buys", "taq_retimb"]].describe().T
print(summary_stats)
```

```python
# number of unique tickers in rh_herd
num_unique_tickers = rh_herd["ticker"].nunique()
print(f"Number of unique tickers in rh_herd: {num_unique_tickers}")
```
