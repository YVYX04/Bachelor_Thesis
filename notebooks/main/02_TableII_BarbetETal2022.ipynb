In this notebook, the purpose is to reproduce Table II in Barber et al. (2022). The table depicts the summary statistics of the herding events.

## 1. Libraries & Data

I first load the relevant libraries and data.

```python
# libs 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import seaborn as sns
from datetime import datetime

# data
rh_herd = pd.read_csv("../../data/processed/CRSP_RH_TAQ_herding_events_only.csv")

# parses dates
rh_herd["date"] = pd.to_datetime(rh_herd["date"], format="%Y-%m-%d")
```

Once I properly loaded the data, I build all the needed variables.

```python
# build userchg
rh_herd["userchg"] = rh_herd["users_close"] - rh_herd["users_close_lag1"]

# make sure that the 'prc' is absolute
rh_herd["prc"] = rh_herd["prc"].abs()

# build size 
rh_herd["size"] = rh_herd["prc"] * rh_herd["shrout"] / 1e6 # size in millions

# build 'daily_buys'
rh_herd["daily_buys"] = rh_herd["buy_num_trades_retail"]

# build 'daily_sells'
rh_herd["daily_sells"] = rh_herd["sell_num_trades_retail"]

# build 'net_buys'
rh_herd["net_buys"] = rh_herd["daily_buys"] - rh_herd["daily_sells"]

# build taq_retimb = net_buys / (daily_buys + daily_sells)
rh_herd["taq_retimb"] = rh_herd["net_buys"] / (rh_herd["daily_buys"] + rh_herd["daily_sells"])
```

Once all the variables are built, I can print the summary statistics.

```python
# summary stats
summary_stats = rh_herd[["users_close", "users_last", "userchg", "userratio", "prc", "size", "ret", "daily_buys", "daily_sells", "net_buys", "taq_retimb"]].describe().T
print(summary_stats)
```
